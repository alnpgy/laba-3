```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

url = "https://conomy.ru/education/articles/147"
id_table = 0
df_power = pd.read_html(url, header = 0, index_col = 0)[id_table]
df_power
```

```python
df_power['Общая мощность всех видов выработки электроэнергии'] = df_power['«Интер РАО»']+df_power['«РусГидро»']+df_power['«Юнипро»']+df_power['«Мосэнерго»']+df_power['ОГК-2']+df_power['ТГК-1']+df_power['«Энел Россия»']
df_power
```

```python
df_power = df_power.sort_values(['Общая мощность всех видов выработки электроэнергии'], ascending = False)
df_power
```

```python
print(df_power[df_power['Общая мощность всех видов выработки электроэнергии'] < 3333])
print()
```

```python
df_power_2 = df_power.copy()
print(df_power_2.iloc[3 , 3])
print()
print(df_power_2.iloc[: -3, 2 : 4])
print()
print(df_power_2.iloc[ : : 1, 2])
print()
```

```python
print(df_power.groupby(['Общая мощность всех видов выработки электроэнергии']).max())
print()
```

```python
print(df_power.groupby(['ТГК-1', df_power['Общая мощность всех видов выработки электроэнергии'] > 300])['Общая мощность всех видов выработки электроэнергии'].count())
print()
```

```python
df_power['«Интер РАО»'].plot(kind = 'bar', color='green')
df_power['«РусГидро»'].plot(kind = 'bar', color='white')
df_power['«Юнипро»'].plot(kind = 'bar', color='blue')
df_power['«Мосэнерго»'].plot(kind = 'bar', color='pink')
df_power['ОГК-2'].plot(kind = 'bar', color='orange')
df_power['ТГК-1'].plot(kind = 'bar', color='yellow')
df_power['«Энел Россия»'].plot(kind = 'bar', color='black')
print()
```

```python
input_matrix = df_power.values
generation_sum = (input_matrix[: , :7 ]).sum(axis = 0)

labels = df_power.columns[ : 7]

colors = ['lightblue', 'gold', 'mediumseagreen', 'lightskyblue', 'coral', 'orchid', 'lightgray']
fig1, ax1 = plt.subplots(figsize = (5, 5))
ax1.pie(generation_sum, labels=labels, autopct='%0.1f%%', colors=colors)
ax1.axis('equal')
plt.show()
```
